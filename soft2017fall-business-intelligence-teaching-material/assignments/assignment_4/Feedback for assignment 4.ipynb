# Plotting Histograms

When you plot a cummulative histogram, for data where you want to split on frequency values (typically y-axis). Then the values on the x-axis cannot change for the same dataset. See the example below.

````python
%matplotlib notebook


freq, vals = np.histogram(prices, bins='auto')
plt.bar(vals[:-1], freq, width=500, linewidth=0, align='center')
plt
````

````python
%matplotlib notebook

# just to generate some example splits of data
freqs = []
for f in freq:
    fst_bound = f // 2
    snd_bound = fst_bound // 3
    freqs.append((fst_bound, snd_bound, f - (fst_bound + snd_bound)))

freqs_low, freqs_medium, freqs_high = zip(*freqs)
plt.bar(vals[:-1], freqs_low, width=500, linewidth=0, align='center')
plt.bar(vals[:-1], freqs_medium, width=500, linewidth=0, align='center', bottom=freqs_low)
plt.bar(vals[:-1], freqs_high, width=500, linewidth=0, align='center', bottom=freqs_medium)
plt
````

## Redefining Keywords

It is not a good idea to redefine keywords in Python. That is, a variable should not be called `zip` as there is a function with that name. For example, the `zip` function zips collections into tuples.

````python
for el in zip([1, 2, 3], ['a', 'b', 'c']):
    print(el)
````

When you redefine `zip` then your program might behave unexpectetly. For example, the following prints `2200` as expected

````python
zip = 2200
print(zip)
````

However, now following code raises an exception.

````python
zip = 2200

for el in zip([1, 2, 3], ['a', 'b', 'c']):
    print(el)
````
